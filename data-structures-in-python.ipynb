# Data Structures in Python
Data structures are the smallest building blocks in the programming process. They allow us to hold different types of data that represent the situations we, as users, want to tell computers. Data types are divided into Primitive and Non-Primitive



```python
# It is the basic data type in Python that represents whole numbers, and it is the most used data type.

x = 25
type(x)
```

```python
# Represents decimal numbers

x = 2.3
type(x)
```

```python
# Represents textual expressions. The string expression is specified between two apostrophes or quotation marks.

x = 'Hello_AI'
type(x)
```

```python
# Represents two values: True and False.

a = True
print(type(a))
```

```python
list = ['w', 'y', 1, 2, True, [3, 4, 5]] 
type(list)
```

```python
dictionary = {"REG": "Regression",
              "LOG": "Logistic Regression",
              "CART": "Classification and Reg"}
type(dictionary)
```

```python
t = ("apple", "green", 1, 2)
type(t)
```

```python
set = {2, 3, 4, "x", "y"}
type(set)
```
